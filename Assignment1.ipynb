```
import neuralnets as nn
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1, 1, 200)[:,None]
y = x**3 + x**2 + np.random.normal(0, 0.2, (200,1))
plt.scatter(x, y)
plt.show() 
class architecture(nn.Module):
  def __init__(self):
    super().__init__()
    weight = nn.init.RandomUniform()
    bias = nn.init.Constant(0.1)
    self.l1 = nn.layers.Dense(1, 8, nn.act.tanh, weight, bias)
    self.l2 = nn.layers.Dense(8, 8, nn.act.tanh, weight, bias)
    self.out = nn.layers.Dense(8, 1, w_initializer = weight, b_initializer = bias)
  
  def forward(self,x):
    x = self.l1(x)
    x = self.l2(x)
    o = self.out(x)
    return o

obj = architecture()
optimize = nn.optim.Adam(obj.params, lr = 0.1)
loss_fn = nn.losses.MSE()
l = []
iteration = []
for i in range(100):
    o = obj.forward(x)
    loss = loss_fn(o, y)
    l.append(loss.data)
    iteration.append(i)
    obj.backward(loss)
    optimize.step()
    plt.scatter(x,y)
    plt.plot(x, o.data, c = "green", lw = 3)
    plt.show()
    print(loss.data)

plt.plot( iteration, l )
plt.show()
```
